```python
# one hot encoding
central_toronto_one = pd.get_dummies(torontocity_venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add the postal, borough and neighborhood columns back to the dataframe
central_toronto_one['PostalCode'] = torontocity_venues_df['PostalCode'] 
central_toronto_one['Borough'] = torontocity_venues_df['Borough'] 
central_toronto_one['Neighbourhoods'] = torontocity_venues_df['Neighbourhood'] 

# move the postal, borough and neighborhood columns to the first column in the table
fixed_columns = list(central_toronto_one.columns[-3:]) + list(central_toronto_one.columns[:-3])
central_toronto_one = central_toronto_one[fixed_columns]

print(central_toronto_one.shape)
central_toronto_one.head()
```

```python
central_toronto_venue_frequency = central_toronto_one.groupby(['PostalCode', 'Borough', 'Neighbourhoods']).mean().reset_index()
print(central_toronto_venue_frequency.shape)
central_toronto_venue_frequency.head()
```

```python
top_venue_number = 10

indicators = ['st', 'nd', 'rd']

#use code to create columns according to the number of top venues
areaColumns = ['PostalCode', 'Borough', 'Neighbourhoods']
frequencyColumns = []
for ind in np.arange(top_venue_number):
    try:
        frequencyColumns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        frequencyColumns.append('{}th Most Common Venue'.format(ind+1))
columns = areaColumns+frequencyColumns
#use code to create a new dataframe
sorted_neighborhood_venues = pd.DataFrame(columns=columns)
sorted_neighborhood_venues['PostalCode'] = central_toronto_venue_frequency['PostalCode']
sorted_neighborhood_venues['Borough'] = central_toronto_venue_frequency['Borough']
sorted_neighborhood_venues['Neighbourhoods'] = central_toronto_venue_frequency['Neighbourhoods']

for ind in np.arange(central_toronto_venue_frequency.shape[0]):
    row_categories = central_toronto_venue_frequency.iloc[ind, :].iloc[3:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    sorted_neighborhood_venues.iloc[ind, 3:] = row_categories_sorted.index.values[0:top_venue_number]

sorted_neighborhood_venues.sort_values(frequencyColumns, inplace=True)
sorted_neighborhood_venues
```

```python
central_toronto_venue_frequency.rename(columns= {'PostalCode' : 'PostalCODE'}, inplace = True)
central_toronto_venue_frequency.rename(columns= {'PostalCODE' : 'PostalCode'}, inplace = True)
central_toronto_df.rename(columns={'PostalCODE' : 'PostalCode'}, inplace = True)
central_toronto_df.head()
```

```python
kclusters = 6

central_toronto_venue_frequency_clustering = central_toronto_venue_frequency.drop(['PostalCode', 'Borough', 'Neighbourhoods'], 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(central_toronto_venue_frequency_clustering)

central_toronto_clustered_df = central_toronto_df
central_toronto_clustered_df['Cluster'] = kmeans.labels_

central_toronto_clustered_df = central_toronto_clustered_df.join(sorted_neighborhood_venues.drop(['Borough', 'Neighbourhoods'], 1).set_index('PostalCode'), on='PostalCode')
central_toronto_clustered_df.sort_values(['Cluster'] + frequencyColumns, inplace=True)
central_toronto_clustered_df
```
